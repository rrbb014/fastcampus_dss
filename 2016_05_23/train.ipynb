```python
tt = np.linspace(-100, 500, 100)
F = tt / 360
F[tt < 0]  = 0
F[tt > 360] = 1
plt.plot(tt,F)
plt.ylim(-0.1, 1.1)
plt.xticks([0,180,360])
plt.title("Cumulative Distribution Function");
plt.xlabel("$x$ (deg.)");
plt.ylabel("$F(x)$");
```
