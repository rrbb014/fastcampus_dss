```python
import MySQLdb
import pandas as pd
import numpy as np
```

```python
db = MySQLdb.connect(
    "db.fastcamp.us",
    "root",
    "dkstncks",
    "sakila",
    charset='utf8'
)
```

```python
film_df = pd.read_sql("SELECT * FROM film;",db)
film_df.head(1)
```

```python
is_year = film_df.release_year.isin([2006,2007])
is_grade = film_df.rating.isin(["PG","G"])

film_df[is_year & is_grade]
```

```python
is_099 = film_df.rental_rate == 0.99
is_boring = film_df.description.str.contains("Boring")

film_df[is_099 & is_boring][["title","description","rental_rate"]]
```

```python
film_group = film_df.groupby("rating")
film_group.agg({"film_id": np.size,"rental_rate":np.mean})
```
